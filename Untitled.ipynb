```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import L2
from sklearn.metrics import mean_squared_error
```

```python
# Load the dataset
data = pd.read_csv('data/processedX.csv')
```

```python
# Extract temperature and humidity columns
temp_columns = [col for col in data.columns if 'temperature_2m' in col]
hum_columns = [col for col in data.columns if 'relative_humidity_2m' in col]
```

```python
# Normalize the data
scaler_temp = MinMaxScaler()
scaler_hum = MinMaxScaler()
scaled_temp = scaler_temp.fit_transform(data[temp_columns])
scaled_hum = scaler_hum.fit_transform(data[hum_columns])
scaled_data = np.hstack((scaled_temp, scaled_hum))
```

```python
# Create sequences for time series forecasting
def create_sequences(data, lookback=30):
    sequences = []
    targets = []
    for i in range(len(data) - lookback):
        sequences.append(data[i:i + lookback])
        targets.append(data[i + lookback])
    return np.array(sequences), np.array(targets)
```

```python
# Create sequences
X, y = create_sequences(scaled_data, lookback=30)
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
# Define a more complex LSTM model
model = Sequential([
    Bidirectional(LSTM(256, activation='relu', return_sequences=True,
                       input_shape=(X_train.shape[1], X_train.shape[2]), kernel_regularizer=L2(0.001))),
    Dropout(0.4),
    Bidirectional(LSTM(256, activation='relu', return_sequences=True, kernel_regularizer=L2(0.001))),
    Dropout(0.4),
    Bidirectional(LSTM(128, activation='relu')),
    Dropout(0.3),
    Dense(128, activation='relu'),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dense(14)  # 7 temperature + 7 humidity predictions
])
```

```python
# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), loss='mse')
```

```python
# Callbacks for early stopping and learning rate reduction
early_stopping = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=10, min_lr=1e-6)
```

```python
# Train the model with a larger batch size (128)
history = model.fit(X_train, y_train, epochs=120, batch_size=64, validation_data=(X_test, y_test),
                    callbacks=[early_stopping, reduce_lr])
```

```python
# Save the trained model
model.save('lstm_temperature_humidity_modelv64.h5')  # Save the model to a file
model.save('lstm_temperature_humidity_modelv64.keras')  # Save the model to a file
```

```python
# Evaluate the model
loss = model.evaluate(X_test, y_test)
print(f"Test Loss: {loss:.4f}")
```

```python
# Make predictions
predictions = model.predict(X_test)
```

```python
# Reverse scaling to get actual temperature and humidity values
predicted_temp = scaler_temp.inverse_transform(predictions[:, :7])
predicted_hum = scaler_hum.inverse_transform(predictions[:, 7:])
actual_temp = scaler_temp.inverse_transform(y_test[:, :7])
actual_hum = scaler_hum.inverse_transform(y_test[:, 7:])
```

```python
# Calculate RMSE for temperature and humidity
rmse_temp = np.sqrt(mean_squared_error(actual_temp, predicted_temp))
rmse_hum = np.sqrt(mean_squared_error(actual_hum, predicted_hum))
```

```python
# Calculate accuracy percentage
accuracy_temp = (1 - rmse_temp / np.mean(actual_temp)) * 100
accuracy_hum = (1 - rmse_hum / np.mean(actual_hum)) * 100

print(f"Temperature RMSE: {rmse_temp:.2f}")
print(f"Humidity RMSE: {rmse_hum:.2f}")
print(f"Temperature Accuracy: {accuracy_temp:.2f}%")
print(f"Humidity Accuracy: {accuracy_hum:.2f}%")
```

```python
# Plot training & validation loss
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
import numpy as np
from datetime import datetime, timedelta
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler

# Load the saved model
model = tf.keras.models.load_model('lstm_temperature_humidity_modelv64.keras')

# Function to predict future weather conditions (next 14 days)
def predict_future(model, initial_sequence, future_steps=14):
    future_predictions_temp = []
    future_predictions_hum = []
    current_sequence = initial_sequence
    
    for _ in range(future_steps):
        prediction = model.predict(np.expand_dims(current_sequence, axis=0))[0]  # Predict for the next time step
        
        predicted_temp = prediction[:7]  # First 7 values for temperature
        predicted_hum = prediction[7:]   # Next 7 values for humidity
        
        future_predictions_temp.append(predicted_temp)
        future_predictions_hum.append(predicted_hum)
        
        # Append prediction to the current sequence and remove the first step to create a rolling window
        current_sequence = np.vstack((current_sequence[1:], prediction))
    
    return np.array(future_predictions_temp), np.array(future_predictions_hum)

# Assuming you have loaded your test data (X_test) and scalers for temperature and humidity
# Get the last sequence from the test set to start future prediction
initial_sequence = X_test[-1]  # Last sequence from the test set

# Get today's date and set the time to 12:00 PM (noon)
today = datetime.today().replace(hour=12, minute=0, second=0, microsecond=0)

# Create future dates for the next 14 days at 12:00 PM
future_dates = [today + timedelta(days=i) for i in range(1, 15)]

# Predict the next 14 days
future_predicted_temp, future_predicted_hum = predict_future(model, initial_sequence, future_steps=14)

# Reverse scaling to get actual temperature and humidity values for the future
future_predicted_temp = scaler_temp.inverse_transform(future_predicted_temp)
future_predicted_hum = scaler_hum.inverse_transform(future_predicted_hum)

# Create a DataFrame for future predictions
future_predictions_df = pd.DataFrame({
    'Date (12:00 PM)': future_dates,
    'Predicted Temperature (°C)': future_predicted_temp[:, 0],  # Display first temperature value
    'Predicted Humidity (%)': future_predicted_hum[:, 0]        # Display first humidity value
})

# Display the table with dates and predictions
print(future_predictions_df)

# Optionally save the table to a CSV file
future_predictions_df.to_csv('future_predictions_14_days.csv', index=False)
```
